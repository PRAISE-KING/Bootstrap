```python
import pandas
dataset = pandas.read_csv("https://modcom.co.ke/data/datasets/Advertising.csv")
dataset
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
subset = dataset[["TV", "Radio", "Newspaper", "Sales"]] 
#subset - introduced to reduce number of columns to be used.in this case "No."" was unnecessary.can also use other names e.g y,z or praiz etc
subset
```

```python
array = subset.values
```

```python
X = array[:,0:3]
Y = array[:, 3]
Y
```

```python
dataset.values
```

```python
from sklearn import model_selection #model_selection used to split data into training sets(70%) and testing sets(30%)

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=40)
```

```python
# algorithms

from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
```

```python
model = KNeighborsRegressor()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print("The model predicted : ", predictions)
```

```python
print("The correct answer is :", Y_test)
```

```python
# evaluation - the closeness of model prediction to the real answer (r2_score)

from sklearn.metrics import r2_score
print("the accuracy is : ", r2_score(predictions, Y_test)*100)
```
